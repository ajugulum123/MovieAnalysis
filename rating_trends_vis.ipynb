```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

m1930 = pd.read_csv('export.csv')
m1950 = pd.read_csv('export (1).csv')
m1970 = pd.read_csv('export (2).csv')
m1990 = pd.read_csv('export (3).csv')
m2010 = pd.read_csv('export (4).csv')

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Function to plot scatter plots alongside coefficient information on Rating vs. Number of votes
def plot_with_correlation(ax, df, title, colour):
    ax.scatter(df['Rating'], df['NumVotes'], color=colour, alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Number of votes')
    ax.set_xlim(0, 10)
    ax.grid(True)
    
    corr_coeff, _ = pearsonr(df['Rating'], df['NumVotes'])
    
    ax.text(0.05, 0.95, f'Corr: {corr_coeff:.2f}', transform=ax.transAxes,
            fontsize=10, ha='left', va='top', bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.5))

    x = df['Rating']
    y = df['NumVotes']
    fit = np.polyfit(x, y, 1)  
    ax.plot(x, fit[0] * x + fit[1], color='black', linewidth=2, linestyle='--')
    ax.legend() 

# Scatter plot for 1930-1949 DataFrame
plot_with_correlation(axs[0, 0], m1930, '1930-1949: Number of Votes vs. Rating', 'blue')

# Scatter plot for 1950-1969 DataFrame
plot_with_correlation(axs[0, 1], m1950, '1950-1969: Number of Votes vs. Rating', 'red')

# Scatter plot for 1970-1989 DataFrame
plot_with_correlation(axs[0, 2], m1970, '1970-1989: Number of Votes vs. Rating', 'green')

# Scatter plot for 1990-2009 DataFrame
plot_with_correlation(axs[1, 0], m1990, '1990-2009: Number of Votes vs. Rating', 'purple')

# Scatter plot for 2010-present DataFrame
plot_with_correlation(axs[1, 1], m2010, '2010-present: Number of Votes vs. Rating', 'orange')

# Hide the empty subplot
axs[1, 2].axis('off')

plt.tight_layout()  # Adjust spacing between plots
plt.show()
#fig.savefig('votesvsrating.png')
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Function to plot scatter plots alongside coefficient information on Rating vs. Runtime
def plot_with_correlation(ax, df, title, colour):
    ax.scatter(df['Rating'], df['Runtime'], color=colour, alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Runtime in minutes')
    ax.set_xlim(0, 10)
    ax.grid(True)
    
    corr_coeff, _ = pearsonr(df['Rating'], df['Runtime'])
    
    ax.text(0.05, 0.95, f'Corr: {corr_coeff:.2f}', transform=ax.transAxes,
            fontsize=10, ha='left', va='top', bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.5))

    x = df['Rating']
    y = df['Runtime']
    fit = np.polyfit(x, y, 1)  
    ax.plot(x, fit[0] * x + fit[1], color='black', linewidth=2, linestyle='--')
    ax.legend()  
    
# Scatter plot for 1930-1949 DataFrame
plot_with_correlation(axs[0, 0], m1930, '1930-1949: Runtime vs. Rating', 'blue')

# Scatter plot for 1950-1969 DataFrame
plot_with_correlation(axs[0, 1], m1950, '1950-1969: Runtime vs. Rating', 'red')

# Scatter plot for 1970-1989 DataFrame
plot_with_correlation(axs[0, 2], m1970, '1970-1989: Runtime vs. Rating', 'green')

# Scatter plot for 1990-2009 DataFrame
plot_with_correlation(axs[1, 0], m1990, '1990-2009: Runtime vs. Rating', 'purple')

# Scatter plot for 2010-present DataFrame
plot_with_correlation(axs[1, 1], m2010, '2010-present: Runtime vs. Rating', 'orange')

axs[1, 2].axis('off')

plt.tight_layout()  
plt.show()
#fig.savefig('runtimevsrating.png')
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Function to create histogram of rating distribution 
def plot_histogram(ax, df, title, colour):
    n, bins, patches = ax.hist(df['Rating'], bins=20, color=colour, alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Frequency')
    ax.set_xlim(0, 10)
    ax.grid(True)
    
    mode_rating = df['Rating'].mode().values[0]
    
    ax.text(0.05, 0.95, f'Mode: {mode_rating}', transform=ax.transAxes,
            fontsize=10, ha='left', va='top', bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.5))

# Histogram for 1930-1949 DataFrame
plot_histogram(axs[0, 0], m1930, '1930-1949: Distribution of Ratings', 'blue')

# Histogram for 1950-1969 DataFrame
plot_histogram(axs[0, 1], m1950, '1950-1969: Distribution of Ratings', 'red')

# Histogram for 1970-1989 DataFrame
plot_histogram(axs[0, 2], m1970, '1970-1989: Distribution of Ratings', 'green')

# Histogram for 1990-2009 DataFrame
plot_histogram(axs[1, 0], m1990, '1990-2009: Distribution of Ratings', 'purple')

# Histogram for 2010-present DataFrame
plot_histogram(axs[1, 1], m2010, '2010-present: Distribution of Ratings', 'orange')

axs[1, 2].axis('off')

plt.tight_layout() 
plt.show()
#fig.savefig('ratingdist.png')
```

```python
allmovies = pd.merge(m1930, m1950, how='outer')
allmovies = allmovies.merge(m1970, how='outer')
allmovies = allmovies.merge(m1990, how='outer')
allmovies = allmovies.merge(m2010, how='outer')

allmovies['Year'] = pd.to_datetime(allmovies['Year'], format='%Y')

avg_rating_per_year = allmovies.groupby(allmovies['Year'].dt.year)['Rating'].mean()

# Plotting line chart of average rating over the years 
plt.figure(figsize=(12, 6))
avg_rating_per_year.plot(kind='line', marker='o', color='b', linewidth=2)

plt.title('Average Movie Rating Per Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.xticks(rotation=45)  

plt.tight_layout()
plt.show()
#plt.savefig('ratingtime.png')

fig_save = plt.figure(figsize=(12, 6))
ax_save = fig_save.add_subplot(111)
avg_rating_per_year.plot(kind='line', marker='o', color='b', linewidth=1, ax=ax_save)

ax_save.set_title('Average Movie Rating Per Year')
ax_save.set_xlabel('Year')
ax_save.set_ylabel('Average Rating')
ax_save.grid(True)

# Save the plot as an image
#fig_save.savefig('average_rating_per_year.png')
```

```python
avg_runtime_per_year = allmovies.groupby(allmovies['Year'].dt.year)['Runtime'].mean()

# Plotting line chart of average runtime over the years
plt.figure(figsize=(12, 6))
avg_runtime_per_year.plot(kind='line', marker='o', color='g', linewidth=2)

plt.title('Average Movie Runtime Per Year')
plt.xlabel('Year')
plt.ylabel('Average Runtime (minutes)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout()

# Show the plot
plt.show()
#plt.savefig('runtimetime.png')

# Create a new figure instance for saving
fig_save2 = plt.figure(figsize=(12, 6))
ax_save2 = fig_save2.add_subplot(111)
avg_runtime_per_year.plot(kind='line', marker='o', color='g', linewidth=1, ax=ax_save2)

ax_save2.set_title('Average Movie Runtime Per Year')
ax_save2.set_xlabel('Year')
ax_save2.set_ylabel('Average Runtime (minutes)')
ax_save2.grid(True)

# Save the plot as an image
#fig_save2.savefig('average_runtime_per_year.png')
```
